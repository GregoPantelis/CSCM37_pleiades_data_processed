# Design 4

## For this design the 'places' csv file was used.

#### Counting the number of ancient places each Author is involved in

```python
creators = list(p_places.creators.values)
creators_u = list(set(creators))

creators_unique = []

for i in range(len(creators_u)):
    a = creators_u[i].split(", ")
    for ii in range(len(a)):
        creators_unique.append(a[ii])
        
creators_unique = list(set(creators_unique))

counter = []
for crt in creators_unique:
    c = 0
    for i in range(len(creators)):
        if crt in creators[i]:
            c+=1
    counter.append(c)

creators_df = pd.DataFrame(data={'creators': creators_unique,
                                 'frequency': counter})

creators_df

creators_df2 = creators_df[creators_df['frequency']>500]
creators_df2
```

#### Counting the Frequency of every Ancient Place Type in the dataset

```python
feature_typess = list(p_places.featureTypes.values)
feature_types = list(set(feature_typess))
feature_types

feature_types_unique = []

for i in range(len(feature_types)):
    a = feature_types[i].split(", ")
    for ii in range(len(a)):
        feature_types_unique.append(a[ii])
        
feature_types_unique = list(set(feature_types_unique))

counter = []
for crt in feature_types_unique:
    c = 0
    for i in range(len(feature_typess)):
        if crt in feature_typess[i]:
            c+=1
    counter.append(c)
    
feature_types_df = pd.DataFrame(data={'feature types': feature_types_unique,
                                 'frequency': counter})

feature_types_df
```

#### Getting every Type with frequency above 100

```python
f_t_100 = feature_types_df[feature_types_df['frequency']>100].reset_index(drop=True)
f_t_100
```

```python
f_t_100.to_csv('f_t_100_des4.csv')
```

#### Constracting the data for the desing.
#### We are counting the frequency of involvments of every Authroe for every feature type (with count above 100)

```python
creators2 = list(creators_df2.creators.values)
feat_t = list(f_t_100['feature types'].values)


result = []

for creator in creators2:
    #print(creator)
    c_list = []
    for feat in feat_t:
        c=0
        for i in range(len(p_places)):
            if creator in p_places['creators'][i]:
                if feat in p_places['featureTypes'][i]:
                    c+=1
        c_list.append(c)
    result.append(c_list)
    
a = result
a = np.array(a).T.tolist()

creators_df2 = creators_df2.reset_index(drop=True)
des3_df = creators_df2

for i, feat in enumerate(feat_t):
    des3_df[feat] = a[i]
    
des3_df
```

# Building the Desing 4

```python
des3_df = pd.read_csv('des3_df.csv', index_col=0)
f_t_100 = pd.read_csv('f_t_100_des4.csv', index_col=0)

threshold = pd.DataFrame([{"threshold": 1000}])
selector = alt.selection_multi(fields=['creators'])
feat_t = list(f_t_100['feature types'].values)


########################
## CREATORS BAR CHART ##
########################
bars = alt.Chart(des3_df).mark_bar().encode(
    x=alt.X("frequency:Q", title='Frequency'),
    y=alt.Y("creators:O", title='Authors',
            sort=alt.Sort(field="frequency", order="descending"),
            ),
    color=alt.condition(selector, #brush,
                        'frequency:Q',
                        alt.value('lightgray'),
                        #legend=None
                        legend=alt.Legend( title='No. Involvements')
                       )
).properties(
    title="No. of Ancient Places Involvments per Author",
    #fontSize=2,
    #titleFontSize = 14,
    height=700
).transform_calculate(
    "baseline", "90"
).add_selection(
    selector #brush
)


'''
text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3,  # Nudges text to right so it doesn't appear on top of the bar
   # text=alt.Text('frequency:Q')
).encode(
    text='frequency:O'
)
'''



highlight = alt.Chart(des3_df).mark_bar(color="#e45755").encode(
    x=alt.X('frequency:Q'),
    x2= alt.X2('baseline:Q'),
    y=alt.Y('creators:O', sort=alt.Sort(field="frequency", order="descending")),
).transform_calculate("baseline", "1000").transform_window(
    rank='rank(frequency)',
    sort=[alt.SortField("frequency", order="descending")],
).transform_filter(
    alt.datum.frequency > 1000
).transform_filter(
    selector #brush
)

rule = alt.Chart(threshold).mark_rule().encode(
    x='threshold:Q'
)

rule2 = alt.Chart(des3_df).mark_rule(color='red').encode(
    x='mean(frequency):Q',
)



#######################
##  PARALLEL COORDS  ##
#######################
par_cord = alt.Chart(des3_df).transform_window(
    index='count()'
).transform_fold(
    feat_t
).mark_line().encode(
    y=alt.Y('key:N', title='Types of Ancient Places'),
    x=alt.X('value:Q', title='Frequency'),
    #color='frequency:O',
    color=alt.condition(selector, #brush,
                        'frequency:Q',
                        alt.value('lightgray'),
                        legend=None
                       ),
    detail='index:N',
    opacity=alt.value(0.5)
).properties(
    title="No. Authors' involements in places' types",
    #fontSize=20,
    width=700,
    height=700
).add_selection(
    selector #brush
).transform_filter(
    selector #brush
)



#################
##   Merging   ##
#################
creators = (bars + text + highlight + rule + rule2).properties(width=500)

dashboard1 = alt.hconcat(creators, par_cord, spacing=80,padding={"top":75}, title={"text":"Authors' Involvement in Ancient Places Types",
                                                   "fontSize":23,
                                                   "anchor":"middle", 
                                                   "baseline":"alphabetic"}).configure_axis(
    labelFontSize=13,
    titleFontSize=15
)
dashboard1
```
