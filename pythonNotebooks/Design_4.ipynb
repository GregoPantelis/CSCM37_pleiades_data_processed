#### Loading the des3_periods_v2 dataframe which was constracted for the Design 3

```python
des3_periods_v2 = pd.read_csv('des3_periods_v2.csv', index_col=0)
des3_periods_v2
```

### Get the frequnecy of ancient places in each time period

```python
timePeriods = {'P': [-2600000, -18000],
               'ME':[-18000, -9000],
               'NE':[-9000, -5000],
               'CH':[-5000, -2500],
               'BR':[-2500, -1000],
               'A':[-1000, -550],
               'C':[-550, -330],
               'H':[-330, -30],
               'R':[-30, 300],
               'L':[300, 640],
               'M':[640, 1500],
               'MOD': [1500, 2100]}

periods = ['Paleolithic: [-2600000, -18000]',
            'Mesolithic: [-18000, -9000]',
            'Neolithic: [-9000, -5000]',
            'Chalcolithic: [-5000, -2500]',
            'Bronze: [-2500, -1000]',
            'Archaic: [-1000, -550]',
            'Classical: [-550, -330]',
            'Hellenistic-Republican: [-330, -30]',
            'Roman: [-30, 300]',
            'Late Antique: [300, 640]',
            'Medieval/Byzantine: [640, 1500]',
            'Modern Era: [1500, 2100]']


time_periods_df = pd.DataFrame(data={'Periods':periods,
                                     'Period_key': list(timePeriods.keys()),
                                    })


occ = list(des3_periods_v2.periods.values)
freq = []
for key in list(time_periods_df.Period_key.values):
    freq.append(occ.count(key))

avg_year2 = []
for key in list(timePeriods.keys()):
    a = (timePeriods[key][0] + timePeriods[key][1])/2
    avg_year2.append(int(a))

time_periods_df['Anc_Places_freq'] = freq
time_periods_df['Avg_year'] = avg_year2

time_periods_df
```

```python
des3_periods = pd.read_csv('des3_periods.csv', index_col=0)
des3_periods
```

```python
## countries_df_new_des3.csv is a dataframe constracted for des3
## we load it to get the features
countries_df_new_des3 = pd.read_csv('countries_df_new_des3.csv', index_col=0)
feattt = list(countries_df_new_des3.columns.values)[3:]

result = []
for key in list(timePeriods.keys()):
    #rint(key)
    c_list=[]
    for feat in feattt:
        c=0
        for i in range(len(des3_periods)):
            if key == des3_periods['periods'][i]:
                if feat in des3_periods['featureType'][i]:
                    c+=1
        c_list.append(c)
    result.append(c_list)
    
a = result
a = np.array(a).T.tolist()

time_periods_df
for i, feat in enumerate(feattt):
    time_periods_df[feat] = a[i]
    
time_periods_df
```

```python
## Drop the 'unkown' columns. We gonna use this dataframe for another graph 
time_periods2_df = time_periods_df.drop(columns=['unknown'])
```

```python
time_periods_df.to_csv('time_periods_df.csv')
```

# Constracting the Design 5

```python
import pandas as pd
pd.options.mode.chained_assignment = None

import numpy as np
import altair as alt

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:98% !important; }</style>"))


display(HTML("""
<style>
form.vega-bindings {
  position: absolute;
  right: 270px;
  top: 400px;
}
</style>
"""))


time_periods_df = pd.read_csv('time_periods_df.csv', index_col=0)
time_periods2_df = pd.read_csv('time_periods2_df.csv', index_col=0)
countries_df_new_des3 = pd.read_csv('countries_df_new_des3.csv', index_col=0)

feattt = list(countries_df_new_des3.columns.values)[3:]
feattt2 = [x for x in feattt if x !='unknown']



##################
## BAR SELECTOR ##
##################
color_scale = alt.Scale(
    domain=[
            'Paleolithic: [-2600000, -18000]',
            'Mesolithic: [-18000, -9000]',
            'Neolithic: [-9000, -5000]',
            'Chalcolithic: [-5000, -2500]',
            'Bronze: [-2500, -1000]',
            'Archaic: [-1000, -550]',
            'Classical: [-550, -330]',
            'Hellenistic-Republican: [-330, -30]',
            'Roman: [-30, 300]',
            'Late Antique: [300, 640]',
            'Medieval/Byzantine: [640, 1500]',
            'Modern Era: [1500, 2100]'
    ],
    #range=["#A40126", "#D73027", "#F56D43", "#FDAE61", "#FFE090", "#FFFF99", "#E0F3F9", "#ABD9E9", "#75ADD2", "#4575B5", "#3145ce", "#314553"]
    range=["#481567FF", "#453781FF ", "#39568CFF ", "#2D708EFF ", "#238A8DFF ", "#20A387FF ", "#3CBB75FF ", "#55C667FF ", "#73D055FF  ", "#95D840FF  ", "#DCE319FF ", "#FDE725FF "]
    #range=["#FDE725FF", "#DCE319FF ", "#95D840FF ", "#73D055FF ", "#55C667FF ", "#3CBB75FF ", "#20A387FF ", "#238A8DFF ", "#2D708EFF  ", "#39568CFF  ", "#453781FF ", "#481567FF "]
)


selector = alt.selection_multi(fields=['Avg_year'])

bar_per = alt.Chart(time_periods_df).transform_joinaggregate(
    Total='sum(Anc_Places_freq)',
    
).transform_calculate(
    PercentOfTotal="datum.Anc_Places_freq / datum.Total"
    
).mark_bar().encode(
    alt.Y('PercentOfTotal:Q', axis=alt.Axis(format='.0%'), title='Percentage of Total (Ancient Places)'),
    #y='Avg_year:O'
    x=alt.X('Avg_year:O', title='Average Year', axis=alt.Axis(labelAngle=0),
            sort=alt.EncodingSortField('Avg_year', order='descending')),
   # color=alt.Color(
   #     'Periods:N',
   #     legend=alt.Legend( title='Time Periods'),
   #     scale=color_scale,
   # )
    color=alt.condition(selector,
                       'Periods:N',
                        alt.value('lightgray'),
                        legend=alt.Legend( title='Time Periods'),
                        #legend=None,
                        scale=color_scale,)
).properties(
    title= "Percentage of Ancient Place through Time Periods",
    width=600,
    height=520
).add_selection(
    selector
)


text = alt.Chart(time_periods_df).mark_text(dx=0, dy=-245, color='black').encode(
    y2=alt.Y2('PercentOfTotal:Q', ),
    #x=alt.X('Anc_Places_freq'),
    x=alt.X('Avg_year:O', sort=alt.Sort(field="Avg_year", order="ascending")),
    text=alt.Text('sum(Anc_Places_freq):Q', format='.0f')
).transform_filter(
    selector
)

barplot = bar_per+text

###############
## RECTAGLES ##
###############

features = feattt2
features_dropdown = alt.binding_select(options=features, name='column')
features_select = alt.selection_single(fields=['column'], bind=features_dropdown, init={'column': 'settlement'}, name="Ancient Place Type")


rect = alt.Chart(time_periods_df).transform_fold(
    features,
    as_=['column', 'value']
).transform_filter(
    features_select
).mark_rect().encode(
    alt.X('value:Q',
          bin=True,
          axis=alt.Axis(labelAngle=0, grid=True)),
    alt.Y('Period_key:N', title='Time Periods',
          sort=alt.EncodingSortField('Avg_year', order='descending'),
          axis=alt.Axis(labelAngle=0, grid=True)),
    alt.Color('value:Q',
        scale=alt.Scale(scheme='inferno', reverse=True),
        legend=alt.Legend(title='Total Records')
        ),
    tooltip=['Periods']
).properties(
  #  title= "No. Unknown Type of Ancient Places",
    width=520,
    height=520,
).add_selection(
    features_select
)


#################
## LINE GRAPHS ##
#################
lines1 = alt.Chart(time_periods2_df).transform_window(
    index='count()'
).transform_fold(
    feattt2
).mark_line().encode(
    x=alt.X('key:N', axis=alt.Axis(labelAngle=12)),
    y=alt.Y('value:Q', title='Frequecny'),
    detail='index:N',
    #row=alt.Row('Periods:N', sort=list(cr4.Periods.values)),
    color=alt.Color('Periods',
                    #legend=alt.Legend( title='Time Periods'),
                    legend=None,
                    scale=color_scale,)
                  #  orient='left'),
    #size = alt.Size('value:Q')
    #strokeDash='symbol',
).properties(
    title= "Frequencies of each Feature Type",
    width=1600,
    height=500
).add_selection(
    selector
).transform_filter(
    selector
)

bubble = alt.Chart(time_periods_df).transform_window(
    index='count()'
).transform_fold(
    ['unknown']#feattt2
).mark_circle().encode(
    x=alt.X('key:N', axis=alt.Axis(labelAngle=0)),
    y=alt.Y('value:Q', title='Frequecny'),
   # detail='index:N',
    #row=alt.Row('Periods:N', sort=list(cr4.Periods.values)),
    color=alt.Color('Periods',
                    legend=alt.Legend( title='Time Periods'),
                    scale=color_scale,),
                    #scale=alt.Scale(range=[0, 4000])),
    size = alt.Size('value:Q',
                    scale=alt.Scale(range=[0, 8500]),
                   ),#  legend=alt.Legend( orient='top-left'))
    #strokeDash='symbol',
    tooltip=['Periods', 'unknown']
).properties(
    title= "No. Unknown Type of Ancient Places",
    width=200,
    height=500,
).add_selection(
    selector
).transform_filter(
    selector
)

line_graphs = (lines1 | bubble)
barp_rec = alt.hconcat(barplot, rect, spacing=156)

dashboard3 = alt.vconcat(barp_rec, line_graphs, spacing=20,padding={"top":75}, title={"text":"Types of Ancient Places through Historic Time Periods",
                                                   "fontSize":30,
                                                   "anchor":"middle", 
                                                   "baseline":"alphabetic"}).configure_axis(
    labelFontSize=14,
    titleFontSize=15
).configure_legend(
    titleFontSize=18,
    labelFontSize=15,
    labelLimit= 0,
) #.configure(background='#F8F8F8')
dashboard3
```
