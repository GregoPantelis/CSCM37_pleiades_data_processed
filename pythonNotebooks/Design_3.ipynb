### Getting the Country location of each feature, from the 'locations' csv file
#### It took ~6hours using the geopy library. Results were saved in a csv file.

```python
# GEOLOCATIONS
geocomplete = pd.read_csv('geoComplete.csv', index_col=0)
geocomplete
```

```python
## Appending every feature's country
countr = list(geocomplete.Country.values)
c_code = list(geocomplete.Code.values)

locations_v1 = p_locations[['description', 'featureType', 'pid', 'minDate', 'maxDate']]
locations_v1['country'] = countr
locations_v1['c_code'] = c_code


## Drop Nan values
locations_v2 = locations_v1[~locations_v1['country'].isin(['Nan'])]
locations_v3 = locations_v2[locations_v2['minDate'].notna()]
locations_v3 = locations_v3[locations_v3['country'].notna()]

## Converint minDate and maxDate from str to int
locations_v3 = locations_v3.astype({"minDate": int, "maxDate": int})
locations_v3 = locations_v3.reset_index(drop=True)

## Dropping features with empty feature type (Total: 30 rows)
locations_v3 = locations_v3[locations_v3['featureType'].notna()].reset_index(drop=True)

locations_v3
```

#### Translating every country name to English

```python
import pytz

codes_unique = list(set(list(locations_v3.c_code.values)))

## xk which is Kosovo is no longer available in the pytz libabry,
## so we need to add it manually, thus we need the index
print('Kosovo (kx) index: ', codes_unique.index('xk'))

country_unique = []

for c in codes_unique:
    #print(c)
    if c!='xk':
        country_unique.append(pytz.country_names[c])

country_unique.insert(15, 'Kosovo')

countries_df = pd.DataFrame(data={'Country': country_unique,
                                  'C_code': codes_unique})

countries_df
```

#### Frequency of Ancient Places in each country

```python
freq = []

for c in codes_unique:
    cntr = 0
    for i in range(len(locations_v3)):
        if locations_v3['c_code'][i] == c:
            cntr+=1
    freq.append(cntr)
   
    
countries_df['anc_places'] = freq
countries_df
```

```python
countries_df[countries_df['Country']=='Cyprus']
```

#### Counting the frequency of each Ancient Place type for each country

```python
feat_typs = list(locations_v3.featureType.values)
feat_typ_loc_v3 = list(set(feat_typs))

print(len(feat_typ_loc_v3))

feat_typ_loc_v3_unique = []

for i in range(len(feat_typs)):
    a = feat_typs[i].split(",")
    for ii in range(len(a)):
        feat_typ_loc_v3_unique.append(a[ii])
        
feat_typ_loc_v3_unique = list(set(feat_typ_loc_v3_unique))
print(len(feat_typ_loc_v3_unique))

result = []
#p = 0
for cntry in codes_unique:
    #p+=1
    #print(p)
    c_list = []
    for feat in feat_typ_loc_v3_unique:
        c=0
        for i in range(len(locations_v3)):
            if cntry in locations_v3['c_code'][i]:
                if feat in locations_v3['featureType'][i]:
                    c+=1
        c_list.append(c)
    result.append(c_list)
    
a = result
a = np.array(a).T.tolist()

countries_df_new = countries_df

for i, feat in enumerate(feat_typ_loc_v3_unique):
    countries_df_new[feat] = a[i]
    

countries_df_new
```

#### Get only the columns which have unique items length greater than 7.
#### This is done to get rid of any columns (feature types) that have very low values, their frequnecy is not so important beacuse they appear only a few times

```python
countries_df_new = countries_df_new.drop(columns=[''])
countries_df_new.drop([col for col in countries_df_new.columns if (len(countries_df_new[col].unique()) <=7)],axis=1, inplace=True )
#cr2.to_csv('cr2v2.csv')
countries_df_new
```

### Now we need to process the Time Periods

```python
locations_v3
```

```python
## We translate every country name to engish 

countries = []

for i in range(len(locations_v3)):
    #print(c)
    if locations_v3['c_code'][i] !='xk':
        countries.append(pytz.country_names[locations_v3['c_code'][i]])
    else:
        countries.append('Kosovo')
        
locations_v3['countries'] = countries
locations_v3 = locations_v3.drop(columns=['country'])
locations_v3
```

```python
## Sort the dataframe by min and max date

des3_periods = locations_v3.sort_values(by=['minDate', 'maxDate']).reset_index(drop=True)
des3_periods = des3_periods.drop(columns='description')
des3_periods
```

### Here we definining the range of every time period, and we assin the time period accordingly to every feature.
#### We assign each feature to the proper time period by checking if their min,maxDate average is within the range of the time Period

```python
timePeriods = {'P': [-2600000, -18000],
               'ME':[-18000, -9000],
               'NE':[-9000, -5000],
               'CH':[-5000, -2500],
               'BR':[-2500, -1000],
               'A':[-1000, -550],
               'C':[-550, -330],
               'H':[-330, -30],
               'R':[-30, 300],
               'L':[300, 640],
               'M':[640, 1500],
               'MOD': [1500, 2100]}

periods = []

for i in range(len(des3_periods)):
    a = (des3_periods['minDate'][i]+des3_periods['maxDate'][i])/2
    for key in list(timePeriods.keys()):
        if int(a) in range(timePeriods[key][0],timePeriods[key][1]):
            periods.append(key)

des3_periods['periods'] = periods


avg_year = []          
for i in range(len(des3_periods)):
    for key in list(timePeriods.keys()):
        if key == des3_periods['periods'][i]:
            #a = (timePeriods[key][0]+timePeriods[key][1])/2
            a = (des3_periods['minDate'][i]+des3_periods['maxDate'][i])/2
            avg_year.append(int(a))
    
des3_periods['avg_year'] = avg_year

## Sort based on average year
des3_periods = des3_periods.sort_values(by=['avg_year']).reset_index(drop=True)

des3_periods
```

```python
des3_periods.to_csv('des3_periods.csv')
```
